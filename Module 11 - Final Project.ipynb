#### Tic-Tac-Toe Board

```python
### Tic-Tac-Toe Board

def visualize_board_simple(board):
  ''' 
  A fast an easy visualization of the Tic-Tac-Toe
  gameboard.
  '''
  for row in board:
    print(" ".join(row))


def visualize_board_complicated(board):
  ''' 
  A more complicated, but prettier visualization of the Tic-Tac-Toe
  gameboard.
  '''
  for i in range(len(board)):
    for j in range(len(board[i])):
      board[i][j] = board[i][j].replace("_"," ").replace("o","O").replace("x","X")
  new_board = [["  ", " ", "  |", "  ", " ", "  |", "  ", " ", "  ",],
               ["  ", board[0][0], "  |", "  ", board[0][1], "  |", "  ", board[0][2], "  "],
               ["__", "_", "__|", "__", "_", "__|", "__", "_", "__",],
               ["  ", " ", "  |", "  ", " ", "  |", "  ", " ", "  ",],
               ["  ", board[1][0], "  |", "  ", board[1][1], "  |", "  ", board[1][2], "  "],
               ["__", "_", "__|", "__", "_", "__|", "__", "_", "__",],
               ["  ", " ", "  |", "  ", " ", "  |", "  ", " ", "  ",],
               ["  ", board[2][0], "  |", "  ", board[2][1], "  |", "  ", board[2][2], "  "],
               ["  ", " ", "  |", "  ", " ", "  |", "  ", " ", "  ",]]
  for row in new_board:
    print("".join(row))


current_board = [["_","_","_"],
                 ["_","_","x"],
                 ["_","o","_"]]


print("Simple Visualization")
visualize_board_simple(current_board)

print("\nComplex Visualization, but prettier ")
visualize_board_complicated(current_board)
```

#### Snake

```python
### Snake 
import numpy as np
from pandas import DataFrame

def visualize_snake_board(snake_arr, apple_position, board_dims):
  ''' 
  Uses a pandas Dataframe to evenly display an x,y board for a
  snake game
  '''
  score = len(snake_arr)-1
  x, y = board_dims
  
  head_symbol = "x"
  body_symbol = "·"
  apple_symbol = "A"

  # Make snake and apple symbols
  board = np.full((x,y), " ")
  for key, chunk in enumerate(snake_arr):
    if key == 0:
      board[chunk[0], chunk[1]] = head_symbol
    else:
      board[chunk[0], chunk[1]] = body_symbol
  board[apple_position[0], apple_position[1]] = apple_symbol

  # Make edges
  board = np.vstack([np.full((1,y), "_"), board, np.full((1,y), "-")])
  board = board.T
  board = np.vstack([np.full((1,x+2), "|"), board, np.full((1,x+2), "|")])
  board = board.T

  print("Score:", score)
  print(DataFrame(board))
      

snake_arr = [[5,5],[5,4],[5,3],[5,2],[4,2],[3,2],[3,1],[2,1],[1,1]]
visualize_snake_board(snake_arr = snake_arr, apple_position = [3,8], board_dims = (10,10))
```

#### Mad-Libs

```python
### Mad-Libs

start_bold = '\033[1m'
end_bold = '\033[0;0m'
start_underline = '\033[4m'
end_underline = '\033[0m'
start_italics = '\x1B[3m'
end_italics = '\x1B[0m'

# There is definetly a better way to do this with regex, but 
# I don't know how. Please feel free to correct if you're up to it.
def visualize_madlibs_text(lines):
  for line in lines:
    line = line.replace("**_begin", start_bold + start_underline)
    line = line.replace("**_end", end_bold + end_underline)
    line = line.replace("*_begin", start_italics + start_underline)
    line = line.replace("*_end", end_italics + end_underline)
    print(line)


# lines = ["It was a **droll**, cold Novermber day",
#          "I woke up to the **putrid** smell of **pigeon** roasting downstairs",
#          "I **rolled** down the stairs to see if I could help *verb* the dinner",
#          "My mom said 'See if *relative's name* needs a fresh *noun*'.",
#          "So I carried a tray of glasses full of *a liquid* into the *verb ending in -ing* room.",
#          "When I got there, I couldn't believe my *part of the body (plural)*!",
#          "There were *plural noun* *verb ending in -ing* on the *noun*!"]

lines = ["It was a **_begindroll**_end, cold Novermber day",
         "I woke up to the **_beginputrid**_end smell of **_beginpigeon**_end roasting downstairs",
         "I **_beginrolled**_end down the stairs to see if I could help *_beginverb*_end the dinner",
         "My mom said 'See if *_beginrelative's name*_end needs a fresh *_beginnoun*_end'.",
         "So I carried a tray of glasses full of *_begina liquid*_end into the *_beginverb ending in -ing*_end room.",
         "When I got there, I couldn't believe my *_beginpart of the body (plural)*_end!",
         "There were *_beginplural noun*_end *_beginverb ending in -ing*_end on the *_beginnoun*_end!"]

visualize_madlibs_text(lines)

```
