Let's first recap some material you will need to know for this module...

An easy way to print variables in strings is by using  `f-strings`. `f-strings` are a way to (**f**)ormat `variables` into strings, which make them nice in `print` statements.

To use an `f-string`, you put an `f` right before your quotes/ apostrophes without any space, like `f" my string"` or `f'my string'`. To use a variable in your string, say the variable `x`, you surround it using curly braces (`{}`). For example, you can use `f'x = {x}'` to print out the value of your variable `x`.

````python
n_eggs = 6
sentence = f'The chicken laid {n_eggs} eggs.'
print(sentence)

decimal_num = 1.8964523
print(f'{decimal_num} is approx. {decimal_num:.2f}')
````

The `f` in front of the string tells it to format the variables inside the `{}` into strings. You can also reformat numbers by using `:` afterward. For example, we formatted `decimal_num` into  a float with 2 decimal points (by rounding) using `:.2f`. Here, the `f` specifies a `float`. 

The way we will get input from the user is via the `input` command. For example, we can use

```
name = input("What is your name?")
print(f"Hello {name}! Welcome to my program :)")
```
When the code is run, a text entry box will appear. Simply hit `enter` when you are done typing.

Let's try it!

````python
name = input("What is your name? ")
print(f"Hello {name}! Welcome to my program :)")

# recall f-strings -- give an overview
````

# Group Activity: Let's complete a Madlib together

````python
food = input('A Food or Silly Word: ')
job = input('Profession/Job: ')
adjective = input('Adjective: ')
saying = input('Phrase/Lyrics/Saying: ')
animal = input('Animal: ')
verb = input('Verb: ')
place = input('Place: ')
celebrity = input('Celebrity: ')
buyme = input('Something you would buy: ')
things = input('Things (plural): ')
````

````python
madlib = (f"Hi my name is {celebrity}, but my friends call me {adjective} {food}. " +
          f"My favorite color is the color of {things}, and my favorite thing to do is {verb}. " +
          f"My parents were a {animal} and {job}, which is why we lived in {place}. " +
          f"You probably know me from my TV commercial for {buyme}. " +
          f"I'm the one who says, '{saying}' at the very end!")
print(madlib)
````
